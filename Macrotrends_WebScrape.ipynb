```python
# Create driver session in Chrome browser
driver = webdriver.Chrome()

# Maximize the web page to try to get the elements to have the same spacing for each url
# Tell the driver what url to scrape (Apple income statement in this case)
driver.maximize_window()
driver.get(aapl_is)

# Create table variable to grab the div element id for the table on the web page
table = driver.find_element_by_id("wrapperjqxgrid")
driver.execute_script("window.scrollBy(0, 600);")
# Create a variable to hold the div element id for the scrollbar
scrollbar = driver.find_element_by_id("jqxScrollThumbhorizontalScrollBarjqxgrid")

# Use ActionChains to be able to automate the "drag and drop" of the scrollbar when scraping because
# not all of the information we are trying to scrape is visible without dragging the scrollbar over 
actions = ActionChains(driver)

# Had to adjust the range for some urls because of the way that the longer Quarterly Data elements were
# aligned (Research and Development Expenses was sometimes 1 lin, other times 2 lines, and sometimes 3 lines).
# This resulted in having to adjust how far the to drag the scrollbar when scraping the Facebook urls
for i in range(1, 7):
    actions.drag_and_drop_by_offset(scrollbar, i, 0).perform()
```

```python
# Create soup object to access the outerHTML of the table because we still need the html elements 
# to be able to separately access the contents within the table
soup = BeautifulSoup(table.get_attribute('outerHTML'), "html.parser")

# Create a variable to hold the div element role="columnheader" from the html and then grab only
# the text out of this div element
headers_list = soup.find_all('div', {'role': 'columnheader'})
column_headers=[h.text for h in headers_list]

# Create a variable to hold the div element role="row" from the html and grab the text from the immediate
# children of the div by restricting it through recursive=False, or else the scrape passes over some of the
# div elements in the table
div_list = soup.find_all('div', {'role': 'row'})
data = [[x.text for x in div.find_all('div', recursive=False)] for div in div_list]

# Create a dataframe containing the scraped data and column headers
aapl_is_df = pd.DataFrame(data, columns=column_headers)

# Close the webdriver
driver.close()

aapl_is_df.head()
```
